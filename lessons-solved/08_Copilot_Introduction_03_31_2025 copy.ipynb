# Basic Usage

## 1. Code Completion

Copilot provides suggestions as you type. Let's try some examples:

1. Start typing a function definition:
```python
def calculate_bmi(weight, height):
    # Calculate BMI using weight in kg and height in meters
```

2. Copilot will suggest the implementation. Press Tab to accept or keep typing for different suggestions.

````python
def calculate_bmi(weight, height):
    # Calculate BMI using weight in kg and height in meters
    if height <= 0:
        raise ValueError("Height must be greater than zero.")
    if weight <= 0:
        raise ValueError("Weight must be greater than zero.")
    bmi = weight / (height ** 2)
    return round(bmi, 2)

def bmi_category(bmi):
    # Determine BMI category based on the calculated BMI
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 24.9:
        return "Normal weight"
    elif 25 <= bmi < 29.9:
        return "Overweight"
    else:
        return "Obesity"
def main():
    # Get user input for weight and height
    try:
        weight = float(input("Enter your weight in kg: "))
        height = float(input("Enter your height in meters: "))
    except ValueError:
        print("Invalid input. Please enter numeric values.")
        return

    # Calculate BMI and determine category
    try:
        bmi = calculate_bmi(weight, height)
        category = bmi_category(bmi)
        print(f"Your BMI is {bmi} and you are classified as {category}.")
    except ValueError as e:
        print(e)
        
if __name__ == "__main__":
    main()
````

````python
# Create a scatter plot using matplotlib with the following features:
# - Random data points
# - Custom colors based on point values
# - A trend line
# - Proper labels and title
# - A legend
# - Proper

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the random seed for reproducibility
np.random.seed(42)

# Generate random data
x = np.random.normal(0, 1, 100)
y = 2 * x + np.random.normal(0, 1, 100)

# Calculate the trend line
z = np.polyfit(x, y, 1)
p = np.poly1d(z)

# Create the scatter plot
plt.figure(figsize=(10, 6))

# Create scatter plot with points colored by their y-values
scatter = plt.scatter(x, y, c=y, cmap='viridis', alpha=0.6, label='Data Points')

# Add trend line
plt.plot(x, p(x), 'r-', label=f'Trend line (y = {z[0]:.2f}x + {z[1]:.2f})')

# Customize the plot
plt.title('Random Data with Trend Line', fontsize=14)
plt.xlabel('X Values', fontsize=12)
plt.ylabel('Y Values', fontsize=12)

# Add colorbar
plt.colorbar(scatter, label='Y Value Magnitude')

# Add legend
plt.legend(loc='upper left')

# Add grid
plt.grid(True, alpha=0.3)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ipywidgets import interact, FloatSlider

# Use inline backend instead
%matplotlib inline

def plot_with_slope(slope):
    # Set the random seed for reproducibility
    np.random.seed(42)
    
    # Generate random data
    x = np.random.normal(0, 1, 100)
    y = slope * x + np.random.normal(0, 1, 100)
    
    # Calculate the trend line
    z = np.polyfit(x, y, 1)
    p = np.poly1d(z)
    
    # Clear previous plot
    plt.clf()
    
    # Create the scatter plot
    plt.figure(figsize=(10, 6))
    
    # Create scatter plot with points colored by their y-values
    scatter = plt.scatter(x, y, c=y, cmap='viridis', alpha=0.6, label='Data Points')
    
    # Add trend line
    plt.plot(x, p(x), 'r-', label=f'Trend line (y = {z[0]:.2f}x + {z[1]:.2f})')
    
    # Set y-axis limits
    plt.ylim(-5, 5)
    
    # Customize the plot
    plt.title(f'Random Data with Slope = {slope:.2f}', fontsize=14)
    plt.xlabel('X Values', fontsize=12)
    plt.ylabel('Y Values', fontsize=12)
    
    # Add colorbar
    plt.colorbar(scatter, label='Y Value Magnitude')
    
    # Add legend
    plt.legend(loc='upper left')
    
    # Add grid
    plt.grid(True, alpha=0.3)
    
    # Adjust layout
    plt.tight_layout()

# Create interactive slider
slope_slider = FloatSlider(
    value=2.0,
    min=0.1,
    max=5.0,
    step=0.1,
    description='Slope:',
    continuous_update=False
)

# Display interactive plot
interact(plot_with_slope, slope=slope_slider)
````

````python
def calculate_dosage(weight, age, condition):
    """Calculate medication dosage based on patient parameters."""
    # Validate that weight and age are positive values
    if weight <= 0 or age < 0:
        raise ValueError("Weight and age must be positive")
    
    # Calculate the base dosage as twice the weight
    base_dosage = weight * 2
    
    # Adjust dosage for children under 12 years old
    if age < 12:
        base_dosage *= 0.75
    
    # Increase dosage for severe conditions
    if condition == "severe":
        base_dosage *= 1.5
    
    # Return the final dosage rounded to 2 decimal places
    return round(base_dosage, 2)

# Import pytest for writing and running test cases
import pytest
# Import the calculate_dosage function dynamically using importlib
import importlib

# Dynamically import the calculate_dosage function from the specified module
calculate_dosage = importlib.import_module("lessons-solved.calculate_dosage").calculate_dosage

# Test case for a normal adult with no severe condition
def test_normal_case():
    assert calculate_dosage(50, 30, "normal") == 100.0

# Test case for a child under 12 years old with no severe condition
def test_child_case():
    assert calculate_dosage(20, 10, "normal") == 30.0

# Test case for an adult with a severe condition
def test_severe_condition():
    assert calculate_dosage(50, 30, "severe") == 150.0

# Test case for a child under 12 years old with a severe condition
def test_child_severe_condition():
    assert calculate_dosage(20, 10, "severe") == 45.0

# Test case for edge condition where weight is zero
def test_edge_case_zero_weight():
    with pytest.raises(ValueError):
        calculate_dosage(0, 30, "normal")

# Test case for edge condition where age is negative
def test_edge_case_negative_age():
    with pytest.raises(ValueError):
        calculate_dosage(50, -1, "normal")

# Test case for an invalid condition (not "normal" or "severe")
def test_edge_case_invalid_condition():
    with pytest.raises(KeyError):
        calculate_dosage(50, 30, "unknown")
````

### 截 Exercise Create a function that processes medical image data:
   - Accept a numpy array representing an image
   - Apply basic image processing (normalization, filtering)
   - Return the processed image
   - Include proper error handling

````python
import numpy as np
from scipy.ndimage import gaussian_filter
import matplotlib.pyplot as plt

def process_medical_image(image):
    """
    Process a medical image by normalizing and applying a Gaussian filter.

    Parameters:
        image (numpy.ndarray): Input image as a numpy array.

    Returns:
        numpy.ndarray: Processed image.

    Raises:
        ValueError: If the input is not a numpy array or is empty.
    """
    # Validate input
    if not isinstance(image, np.ndarray):
        raise ValueError("Input must be a numpy array.")
    if image.size == 0:
        raise ValueError("Input image cannot be empty.")

    try:
        # Normalize the image to the range [0, 1]
        normalized_image = (image - np.min(image)) / (np.max(image) - np.min(image))

        # Apply a Gaussian filter for smoothing
        filtered_image = gaussian_filter(normalized_image, sigma=1)

        return filtered_image
    except Exception as e:
        raise RuntimeError(f"An error occurred during image processing: {e}")
    
# Example medical image as a numpy array
image = np.random.randint(0, 256, (256, 256), dtype=np.uint8)

# Process the image
try:
    processed_image = process_medical_image(image)
    print("Image processed successfully.")
    
    # Display the raw and processed images
    plt.figure(figsize=(12, 6))

    # Raw image subplot
    plt.subplot(1, 2, 1)
    plt.imshow(image, cmap='gray')
    plt.title("Raw Image")
    plt.axis("off")

    # Processed image subplot
    plt.subplot(1, 2, 2)
    plt.imshow(processed_image, cmap='gray')
    plt.title("Processed Image")
    plt.axis("off")

    # Show the plots
    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"Error: {e}")
````


### 截 Exercise: Write a class for managing patient records
   - Include methods for adding, updating, and retrieving records
   - Implement data validation
   - Add methods for basic statistical analysis

````python
class PatientRecords:
    """
    A class to manage patient records with methods for adding, updating, retrieving,
    and performing basic statistical analysis.
    """

    def __init__(self):
        # Initialize an empty dictionary to store patient records
        self.records = {}

    def add_record(self, patient_id, name, age, weight, condition):
        """
        Add a new patient record.

        Parameters:
            patient_id (str): Unique identifier for the patient.
            name (str): Name of the patient.
            age (int): Age of the patient.
            weight (float): Weight of the patient in kilograms.
            condition (str): Medical condition of the patient.

        Raises:
            ValueError: If any input is invalid.
        """
        if not patient_id or not isinstance(patient_id, str):
            raise ValueError("Patient ID must be a non-empty string.")
        if not name or not isinstance(name, str):
            raise ValueError("Name must be a non-empty string.")
        if not isinstance(age, int) or age <= 0:
            raise ValueError("Age must be a positive integer.")
        if not isinstance(weight, (int, float)) or weight <= 0:
            raise ValueError("Weight must be a positive number.")
        if not condition or not isinstance(condition, str):
            raise ValueError("Condition must be a non-empty string.")

        self.records[patient_id] = {
            "name": name,
            "age": age,
            "weight": weight,
            "condition": condition,
        }

    def update_record(self, patient_id, **kwargs):
        """
        Update an existing patient record.

        Parameters:
            patient_id (str): Unique identifier for the patient.
            kwargs: Fields to update (e.g., name, age, weight, condition).

        Raises:
            KeyError: If the patient ID does not exist.
            ValueError: If any input is invalid.
        """
        if patient_id not in self.records:
            raise KeyError(f"No record found for patient ID: {patient_id}")

        for key, value in kwargs.items():
            if key in self.records[patient_id]:
                if key == "age" and (not isinstance(value, int) or value <= 0):
                    raise ValueError("Age must be a positive integer.")
                if key == "weight" and (not isinstance(value, (int, float)) or value <= 0):
                    raise ValueError("Weight must be a positive number.")
                self.records[patient_id][key] = value

    def get_record(self, patient_id):
        """
        Retrieve a patient record.

        Parameters:
            patient_id (str): Unique identifier for the patient.

        Returns:
            dict: Patient record.

        Raises:
            KeyError: If the patient ID does not exist.
        """
        if patient_id not in self.records:
            raise KeyError(f"No record found for patient ID: {patient_id}")
        return self.records[patient_id]

    def get_average_age(self):
        """
        Calculate the average age of all patients.

        Returns:
            float: Average age of patients, or 0 if no records exist.
        """
        if not self.records:
            return 0
        total_age = sum(record["age"] for record in self.records.values())
        return total_age / len(self.records)

    def get_average_weight(self):
        """
        Calculate the average weight of all patients.

        Returns:
            float: Average weight of patients, or 0 if no records exist.
        """
        if not self.records:
            return 0
        total_weight = sum(record["weight"] for record in self.records.values())
        return total_weight / len(self.records)

    def get_patient_count(self):
        """
        Get the total number of patients.

        Returns:
            int: Number of patient records.
        """
        return len(self.records)
    
# Create an instance of PatientRecords
records = PatientRecords()

# Add patient records
records.add_record("P001", "John Doe", 30, 70.5, "Diabetes")
records.add_record("P002", "Jane Smith", 25, 60.0, "Hypertension")

# Update a patient record
records.update_record("P001", weight=72.0, condition="Improved")

# Retrieve a patient record
print(records.get_record("P001"))

# Perform statistical analysis
print("Average Age:", records.get_average_age())
print("Average Weight:", records.get_average_weight())
print("Total Patients:", records.get_patient_count())
````

### 截 Exercise: Create a data visualization dashboard
   - Use pandas for data manipulation
   - Create multiple types of plots using matplotlib
   - Add interactive elements

````python
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact, Dropdown

# Sample dataset
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Age": [25, 30, 35, 40, 45],
    "Weight": [55, 70, 68, 80, 60],
    "Height": [160, 175, 170, 180, 165],
    "Condition": ["Healthy", "Diabetes", "Healthy", "Hypertension", "Healthy"]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Function to display a bar chart
def plot_bar_chart(column):
    plt.figure(figsize=(8, 5))
    plt.bar(df["Name"], df[column], color="skyblue")
    plt.title(f"{column} of Patients", fontsize=14)
    plt.xlabel("Name", fontsize=12)
    plt.ylabel(column, fontsize=12)
    plt.xticks(rotation=45)
    plt.grid(axis="y", alpha=0.3)
    plt.tight_layout()
    plt.show()

# Function to display a scatter plot
def plot_scatter_chart(x_column, y_column):
    plt.figure(figsize=(8, 5))
    plt.scatter(df[x_column], df[y_column], color="orange", alpha=0.7)
    plt.title(f"{y_column} vs {x_column}", fontsize=14)
    plt.xlabel(x_column, fontsize=12)
    plt.ylabel(y_column, fontsize=12)
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

# Interactive bar chart
print("Interactive Bar Chart:")
interact(plot_bar_chart, column=Dropdown(options=["Age", "Weight", "Height"], description="Column:"))

# Interactive scatter plot
print("Interactive Scatter Plot:")
interact(plot_scatter_chart, 
         x_column=Dropdown(options=["Age", "Weight", "Height"], description="X-Axis:"),
         y_column=Dropdown(options=["Age", "Weight", "Height"], description="Y-Axis:"))
````
